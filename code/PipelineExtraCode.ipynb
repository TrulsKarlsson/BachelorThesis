# Detection

```python
import os
os.environ["TF_USE_LEGACY_KERAS"] = "1"
from retinaface import RetinaFace
import matplotlib.pyplot as plt

det_faces = RetinaFace.detect_faces("own-pictures/testBildTruls.jpeg")
print(det_faces)

align_face = RetinaFace.extract_faces(img_path = "own-pictures/testBildTruls.jpeg", align = True)
no_align_face = RetinaFace.extract_faces(img_path = "own-pictures/testBildTruls.jpeg", align = False)

f, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))
ax1.imshow(align_face[0])
ax2.imshow(no_align_face[0])
plt.show()
```
