# LFW Dataset Graphs

## ArcFace Graph

```python
import matplotlib.pyplot as plt

# Threshold values
thresholds = [0.5569, 0.6879, 0.8737, 0.9, 0.88, 0.87, 0.75, 0.8, 0.82, 0.84]

# Metric values
metrics = {
    'F1-score': [0.6431478968792401, 0.8095238095238095, 0.8838643371017472, 0.8529698149951315,
                 0.8775510204081632, 0.8833849329205367, 0.8636363636363636, 0.8961748633879781,
                 0.8994594594594595, 0.8917470525187566],
    'Precision': [1.0, 1.0, 0.9090909090909091, 0.8311195445920304, 0.8958333333333334,
                  0.9125799573560768, 1.0, 0.9879518072289156, 0.9788235294117648, 0.9607390300230947],
    'Recall': [0.474, 0.68, 0.86, 0.876, 0.86, 0.856, 0.76, 0.82, 0.832, 0.832],
    'Accuracy': [0.7364729458917836, 0.8396793587174348, 0.8867735470941884, 0.8486973947895792,
                 0.8797595190380761, 0.8867735470941884, 0.8797595190380761, 0.9048096192384769,
                 0.906813627254509, 0.8987975951903807]
}

# Sorting thresholds and rearranging metric values accordingly
sorted_indices = sorted(range(len(thresholds)), key=lambda k: thresholds[k])
thresholds = [thresholds[i] for i in sorted_indices]
for metric_name in metrics:
    metrics[metric_name] = [metrics[metric_name][i] for i in sorted_indices]

max_f1_index = metrics['F1-score'].index(max(metrics['F1-score']))
best_threshold = thresholds[max_f1_index]

plt.figure(figsize=(10, 6))

plt.plot(thresholds, metrics['F1-score'], color='blue', linestyle='-', label='F1-score')

for i, (metric_name, values) in enumerate(metrics.items()):
    if metric_name != 'F1-score':
        linestyle = '--'
        if i == 1:
            color = 'orange'
        elif i == 2:
            color = 'green'
        else:
            color = 'red'
        plt.plot(thresholds, values, color=color, linestyle=linestyle, label=metric_name)

plt.axvline(x=best_threshold, color='black', linestyle='-', label='Best F1-score')

# Title and labels
plt.xlabel('Threshold')
plt.ylabel('Score')
plt.legend()
plt.grid(True)
plt.savefig('ArcFaceGraph.png')
plt.show()
```

```python
import matplotlib.pyplot as plt

# ArcFace data
arcface_thresholds = [0.5569, 0.6879, 0.8737, 0.9, 0.88, 0.87, 0.75, 0.8, 0.82, 0.84]
arcface_f1_scores = [0.6431478968792401, 0.8095238095238095, 0.8838643371017472, 0.8529698149951315,
                     0.8775510204081632, 0.8833849329205367, 0.8636363636363636, 0.8961748633879781,
                     0.8994594594594595, 0.8917470525187566]

# MagFace data
magface_thresholds = [0.6107, 0.5566, 0.67, 0.75, 0.83, 0.8, 0.79, 0.81, 0.78, 0.9]
magface_f1_scores = [0.8876529477196886, 0.851894374282434, 0.931769722814499, 0.968335035750766,
                     0.9609375000000001, 0.9731876861966237, 0.9730807577268196, 0.9693372898120672,
                     0.9688442211055277, 0.8896860986547085]

# AdaFace data
adaface_thresholds = [0.8124, 0.7223, 0.665, 0.9, 0.8, 0.825, 0.81, 0.815, 0.56]
adaface_f1_scores = [0.9457523029682702, 0.9283422459893047, 0.8687782805429864, 0.88,
                     0.9397089397089398, 0.9383215369059656, 0.9389865563598758, 0.942857142857143, 0.6945169712793734]

# Sorting thresholds and rearranging F1-score values accordingly
arcface_sorted_indices = sorted(range(len(arcface_thresholds)), key=lambda k: arcface_thresholds[k])
arcface_thresholds = [arcface_thresholds[i] for i in arcface_sorted_indices]
arcface_f1_scores = [arcface_f1_scores[i] for i in arcface_sorted_indices]

magface_sorted_indices = sorted(range(len(magface_thresholds)), key=lambda k: magface_thresholds[k])
magface_thresholds = [magface_thresholds[i] for i in magface_sorted_indices]
magface_f1_scores = [magface_f1_scores[i] for i in magface_sorted_indices]

adaface_sorted_indices = sorted(range(len(adaface_thresholds)), key=lambda k: adaface_thresholds[k])
adaface_thresholds = [adaface_thresholds[i] for i in adaface_sorted_indices]
adaface_f1_scores = [adaface_f1_scores[i] for i in adaface_sorted_indices]

# Find the threshold where F1-score is highest for each model
arcface_max_f1_index = arcface_f1_scores.index(max(arcface_f1_scores))
arcface_best_threshold = arcface_thresholds[arcface_max_f1_index]

magface_max_f1_index = magface_f1_scores.index(max(magface_f1_scores))
magface_best_threshold = magface_thresholds[magface_max_f1_index]

adaface_max_f1_index = adaface_f1_scores.index(max(adaface_f1_scores))
adaface_best_threshold = adaface_thresholds[adaface_max_f1_index]

# Plotting
plt.figure(figsize=(10, 6))

# Plot ArcFace
plt.plot(arcface_thresholds, arcface_f1_scores, label='ArcFace')#, marker='o')

# Plot MagFace
plt.plot(magface_thresholds, magface_f1_scores, label='MagFace')#, marker='o')

# Plot AdaFace
plt.plot(adaface_thresholds, adaface_f1_scores, label='AdaFace')#, marker='o')

# Plot vertical lines for the thresholds with highest F1-score for each model
plt.axvline(x=arcface_best_threshold, color='blue', linestyle='--', label='ArcFace Best F1-score')
plt.axvline(x=magface_best_threshold, color='orange', linestyle='--', label='MagFace Best F1-score')
plt.axvline(x=adaface_best_threshold, color='green', linestyle='--', label='AdaFace Best F1-score')

# Title and labels
plt.xlabel('Threshold')
plt.ylabel('F1-score')
plt.legend()
plt.grid(True)
plt.savefig('F1-scores.png')
plt.show()
```
